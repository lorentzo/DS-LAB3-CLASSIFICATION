```python
#data structures and operations
import pandas as pd
import numpy as np

#classifiers
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.tree import DecisionTreeClassifier as DTC

#validation
from sklearn.model_selection import cross_val_score as CVs
from sklearn.model_selection import train_test_split

#data pre-processing
from sklearn.decomposition import PCA

#plotting
from matplotlib import pyplot as plt
from sklearn import tree
import graphviz
```

## 1. Importing data
In this laboratory we used seeds dataset from: https://archive.ics.uci.edu/ml/datasets/seeds

```python
#using pandas to read csv file
dataframe = pd.read_csv('seed.txt', delimiter='\t')
dataframe.columns = ["x1", "x2", "x3", "x4", "x5", "x6", "x7", "y"]

print('Classes:', dataframe['y'].unique())
print('Number of examples:', dataframe.shape[0])
print('Number of features:', dataframe.shape[1]-1)

print()
print('Example of data...')
print(dataframe.head())

#creating numpy array of examples and classes
X = dataframe.drop('y',axis=1).values
y = dataframe['y'].values - 1

print()
print('Example of X...')
print(X[1:6])

print()
print('Example of y...')
print(y[1:6])
```

## 2. Training K nearest neighbour classifier

### 2.1. Cross validation, K-folds=5

```python
#knn classifier from sklearn.neighbors
knn = KNN(n_neighbors=3)

#5-fold cross validation
cvs = CVs(knn, X, y, cv=5)
print('Results from 5-fold cross validation:')
print(cvs)
print('Average value of 5-fold cross validation:', np.average(cvs))
```

**NB:** we can see that classifier preforms very well

### 2.2. CV of KNN classifier with different n_neighbour

```python
#nearest neigbour to classify from
nns = range(1,21)
#cross validation on different n_neighbour
avgs = []
for nn in nns:
    knn = KNN(nn)
    avgs.append(np.average(CVs(knn,X,y,cv=5)))

print('Average values of CV from KNN with different nearest neighbour:')
plt.figure(figsize=(15,7))
plt.plot(nns,avgs), plt.plot(nns,avgs,'x') 
plt.xticks(nns), plt.ylabel('Avg CV acc'), plt.xlabel('No of nn')
plt.show()

maxj = np.argmax(avgs)
minj = np.argmin(avgs)
print('Max avg value:', avgs[maxj], 'with K:', nns[maxj])
print('Min avg value:', avgs[minj], 'wiht K:', nns[minj])
```

**NB:** We can see that different n_neighbour don't result in higher avg CV accuracy

### 2.3. CV of KNN with different metrics

```python
#metrics to choose from:
metrics = ['cosine', 'minkowski', 'euclidean', 'manhattan']
metricDict = {1:'cosine', 2:'minkowski', 3:'euclidean', 4:'manhattan'}

#CV of KNN with different metrics:
avgs = []
for m in metrics:
    knn = KNN(n_neighbors=9, metric=m)
    avgs.append(np.average(CVs(knn,X,y,cv=5)))

plt.figure(figsize=(10,5))
plt.bar(range(len(metrics)),avgs)
plt.xticks(range(len(metrics)), metrics)
plt.show()

maxj = np.argmax(avgs)
minj = np.argmin(avgs)
print('Max avg CV value:', avgs[maxj], 'for metric:', metrics[maxj])
print('Min avg CV value:', avgs[minj], 'for metric:', metrics[minj])
```

**NB:** we can see that different distance metric gives simmilar accuracy

### 2.4. CV for KNN with reduced dimensionality (PCA)

```python
#dimensionality reduction
pca = PCA(n_components=2)
pca.fit(X)
X2d = pca.transform(X)

#classifier
knn = KNN(n_neighbors=9)

#cross validaton with reducted dimensinality dataset
cvsALL = CVs(estimator=knn,X=X,y=y,cv=5)
cvs2D = CVs(estimator=knn,X=X2d,y=y,cv=5)

print('Average accuracy for original dimension:', np.average(cvsALL))
print('Average accuracy for 2 dimensions:', np.average(cvs2D))
```

**NB:** we can see that accuracy is simmilar for both dimensions!

## 3. Training Tree Classifier

### Division of dataset to train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, train_size=0.8, random_state=42, shuffle=True)
print('Whole data:', X.shape[0])
print('Train set:', X_train.shape[0], 'proportion to whole:', X_train.shape[0]/X.shape[0])
print('Test set:', X_test.shape[0], 'proportion to whole:', X_test.shape[0]/X.shape[0])
```

### Training Decision classifier from sklearn.tree

```python
#training classifier
dtc = DTC()
dtc.fit(X_train, y_train)

#visualising tree
tree_dot_format = tree.export_graphviz(dtc,out_file=None)
treeImage = graphviz.Source(tree_dot_format)
treeImage
```

### Evaluating tree classifier on test set

```python
yPred = dtc.predict(X_test)
same = (yPred==y_test).sum()
print('accuracy:', same/len(y_test))
```
